```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/cosmetics.csv")

df
```

# **Bar Chart**

```python
# Bar Chart (Matplotlib): Jumlah Produk per Brand
plt.figure(figsize=(20, 5))
plt.bar(df['Brand'].value_counts().index, df['Brand'].value_counts().values, color='pink')
plt.title('Jumlah Produk per Brand')
plt.xlabel('Brand')
plt.ylabel('Jumlah Produk')
plt.xticks(rotation=60)
plt.show()
```

```python
# Bar Chart (Seaborn): Brand dan Label Produknya
plt.figure(figsize=(20, 5))
sns.countplot(x='Brand', data=df, hue='Label')
plt.title('Brand dan Label Produknya')
plt.xlabel('Brand')
plt.ylabel('Jumlah Produk')
plt.xticks(rotation=45)
plt.show()
```

# **Histogram**

```python
# Histogram (Matplotlib): Distribusi Harga Produk
plt.figure(figsize=(8, 5))
plt.hist(df['Price'], bins=20, color='pink', edgecolor='red')
plt.title('Distribusi Harga Produk')
plt.xlabel('Harga Produk')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Histogram (Seaborn): Distribusi Harga Produk
plt.figure(figsize=(8, 5))
sns.histplot(df['Price'], bins=20, kde=True, color='Pink')
plt.title('Distribusi Harga Produk')
plt.xlabel('Harga Produk')
plt.ylabel('Frekuensi')
plt.show()
```

# **Box Plot**

```python
# Boxplot (Matplotlib): Perbandingan Harga Produk Tiap Brand
plt.figure(figsize=(30, 5))
df.boxplot(column='Price', by='Brand', grid=False, patch_artist=True)
plt.title('Perbandingan Harga Produk Tiap Brand')
plt.xlabel('Brand')
plt.ylabel('Harga Produk')
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot (Seaborn): Perbandingan Harga Produk Tiap Brand
plt.figure(figsize=(30, 5))
sns.boxplot(x='Brand', y='Price', hue='Brand', data= df, palette='pastel')
plt.title('Perbandingan Harga Produk Tiap Brand')
plt.xlabel('Brand')
plt.ylabel('Harga Produk')
plt.xticks(rotation=45)
plt.show()
```

# **Scatter Plot**

```python
# Scatter Plot (Matplotlib): Harga vs Peringkat Produk
plt.figure(figsize=(8, 5))
plt.scatter(df['Price'], df['Rank'], c='blue', alpha=0.5)
plt.title('Harga vs Peringkat Produk')
plt.xlabel('Harga Produk')
plt.ylabel('Peringkat Produk')
plt.show()
```

```python
# Scatter Plot (Seaborn): Harga vs Peringkat Produk
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Price', y='Rank', hue='Brand', data=df)
plt.title('Harga vs Peringkat Produk')
plt.xlabel('Harga Produk')
plt.ylabel('Peringkat Produk')
plt.legend(title='Brand')
plt.show()
```

# **INSIGHT DAN TINDAKAN**

**Bar Chart: Jumlah Produk per Brand dan Brand serta Label Produknya**

Insight:
Jumlah Produk per Brand: Menunjukkan brand dengan jumlah produk terbanyak, berarti lebih banyak pilihan untuk pelanggan.
Brand dan Label Produk: Menunjukkan brand yang dominan di kategori tertentu (misalnya skincare atau make-up), menggambarkan variasi produk.

Tindakan yang Diambil:
1. Pemasaran: Fokuskan promosi pada brand dengan produk terbanyak dan sesuaikan dengan kategori produk untuk audiens yang tepat.
2. Stok Produk: Pastikan brand populer selalu tersedia untuk menghindari pelanggan beralih ke brand lain.
3. Diversifikasi Produk: Brand dengan pilihan produk sedikit bisa menambah variasi produk agar lebih banyak pilihan.
4. Influencer: Kolaborasi dengan influencer berdasarkan kategori produk yang sedang tren untuk menarik perhatian audiens lebih besar.

**Histogram: Distribusi Harga Produk**

Insight:
Distribusi harga: Menunjukkan harga produk yang paling banyak muncul, memberikan gambaran tentang rentang harga yang sering dipilih pelanggan.

Tindakan yang Diambil:
1. Penentuan Harga: Fokuskan stok produk pada rentang harga yang paling diminati pelanggan.
2. Strategi Pemasaran: Sesuaikan kampanye untuk produk dengan harga yang sering dipilih pelanggan, baik terjangkau atau premium.
3. Peningkatan Varian Produk: Tambahkan produk di rentang harga yang kurang terisi untuk memperluas pilihan pelanggan.

**Boxplot: Perbandingan Harga Produk Tiap Brand**

Insight:
Perbandingan harga antar brand: Menunjukkan rentang harga tiap brand dan variasinya. Brand dengan harga yang bervariasi menunjukkan banyak pilihan produk.

Tindakan yang Diambil:
1. Strategi Penetapan Harga: Brand dengan rentang harga luas perlu menyesuaikan harga produk agar lebih menarik segmen pasar yang tepat.
2. Segmentasi Pelanggan: Kelompokkan pelanggan berdasarkan preferensi harga (terjangkau vs premium).
3. Pengelolaan Stok dan Pemasaran: Promosikan produk terjangkau untuk audiens tertentu, dan sorot produk premium dengan kualitas terbaik.
4. Perbaikan Produk: Brand harga tinggi tapi kualitas rendah perlu meningkatkan nilai atau fitur produk agar sebanding dengan harga.

**Scatter Plot: Harga vs Peringkat Produk**

Insight:
Hubungan harga dan peringkat produk: Menunjukkan apakah produk dengan harga lebih tinggi cenderung memiliki peringkat lebih baik, atau sebaliknya.

Tindakan yang Diambil:
1. Sesuaikan harga dengan peringkat: Produk harga tinggi tapi peringkat rendah perlu perbaikan kualitas atau diskon.
2. Promosi produk terjangkau dengan kualitas bagus: Produk harga rendah dan peringkat tinggi bisa dipromosikan lebih intens.
3. Rekomendasi produk lebih tepat: Sesuaikan saran produk berdasarkan harga atau peringkat sesuai preferensi pelanggan.
